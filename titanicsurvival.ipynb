```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder 
```

```python
test_df=pd.read_csv('https://raw.githubusercontent.com/NizELITe/Data-Science-and-ML/main/test%20(1).csv')
test_df
```

```python
train_df = pd.read_csv('https://raw.githubusercontent.com/NizELITe/Data-Science-and-ML/main/train%20(1).csv')
train_df
```

```python
survivalclassifier = DecisionTreeClassifier(max_depth = 10)
```

```python
#removing useless data 
train_df.dtypes
```

```python
list = ['Name','PassengerId','Ticket','Cabin']
dftrn=train_df.drop(list,axis=1)
```

```python
dftrn
```

```python
#applying one hot encoding   to convert string into integer

dftrn['Sex'] = pd.get_dummies(dftrn['Sex'], drop_first=True)
dftrn
```

```python
x_train=dftrn
y_train=x_train['Survived']
y_train.head()
```

```python
x_train = x_train.drop(['Survived'],axis=1)
```

```python
#encoding embarked column
le = LabelEncoder()
x_train['Embarked']= le.fit_transform(x_train['Embarked'])
x_train
```

```python
#checking if null values are in dataset
x_train.isnull()
```

```python
#since there are null values so we will replace them with 0
x_train=x_train.fillna(0)
x_train
```

```python
#training the model 
survivalclassifier = survivalclassifier.fit(x_train,y_train)
```

```python
test_df.head()
```

```python
list = ['Name','PassengerId','Ticket','Cabin']
dftest=test_df.drop(list,axis=1)
dftest.head()
```

```python
dftest['Sex']= pd.get_dummies(dftest['Sex'],drop_first=True)
x_test = dftest
x_test.head()
```

```python
x_test['Embarked'] = le.fit_transform(x_test['Embarked'])
x_test = x_test.fillna(0)
x_test.head()
```

```python
predictsurvival= survivalclassifier.predict(x_test)

print('number of survivals: ' ,len(predictsurvival))
```
