```python
#Task 4
import pandas as pd
df4= pd.read_csv("/content/Dataset-SA.csv")
df4

for index, row in df4.iterrows():
    try:
        pass
    except Exception as e:
        print(f"Error processing row {index}: {e}")
        pass
```

```python
df4.head()
```

```python
df4=df4.dropna()
```

```python
df4=df4.drop(["product_name"],axis=1)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split

X = df4['Summary']
y = df4['Sentiment']

tfidf_vectorizer = TfidfVectorizer()
X_tfidf = tfidf_vectorizer.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.2, random_state=42)

classifier = MultinomialNB()
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)
accuracy = classifier.score(X_test, y_test)

print("Accuracy:", accuracy * 100)

new_review = "ok ok product."
new_review_tfidf = tfidf_vectorizer.transform([new_review])
prediction = classifier.predict(new_review_tfidf)

if prediction == 0:
    print("New review is classified as positive.")
elif prediction == 1:
    print("New review is classified as negative.")
else:
    print("New review is classified as neutral.")
```

```python
from sklearn.model_selection import GridSearchCV

parameters = {'alpha': [0.1, 0.5, 1.0]}
grid_search = GridSearchCV(MultinomialNB(), parameters, cv=5, n_jobs=-1)
grid_search.fit(X_train, y_train)

best_classifier = grid_search.best_estimator_
best_prediction = best_classifier.predict(X_test)
best_accuracy = best_classifier.score(X_test, y_test)
print(best_prediction)
print(best_accuracy)
```
